```python
import pandas as pd
import glob,os
import pdb
files = [file for file in glob.glob("03*.csv")]
```

```python
data = []
for file in files:
    data.append(pd.read_csv(file,converters={'fileName':lambda x:str(x)}))
data=pd.concat(data,ignore_index=True)
```

```python
models = pd.unique(data.loc[:,'modelName'])
resData=[]
for m in models:
    cur_data = data[data.loc[:,'modelName']==m]
    hipLens = pd.unique(cur_data.loc[:,'hipLen'])
    for hl in hipLens:
        cur_data2 = cur_data[cur_data.loc[:,'hipLen']==hl]
        weights = pd.unique(cur_data2.loc[:,'weight'])
        for w in weights:
            cur_data3 = cur_data2[cur_data2.loc[:,'weight']==w]
            kneeDirs = pd.unique(cur_data3.loc[:,'kneeDir'])
            for kd in kneeDirs:
                cur_data4 = cur_data3[cur_data3.loc[:,'kneeDir']==kd]
                if len(cur_data4)>1:
                    cur_data4 = cur_data4[cur_data4.constViolat==cur_data4.constViolat.min()]
                    if len(cur_data4)>1:
                        cur_data4 = cur_data4.iloc[[0]]
                resData.append(cur_data4)
resData = pd.concat(resData,ignore_index=True)                  
```
